```python
users_df = clustered_traffic_df[['bs', 'time_hour', 'users']]
users_pivot = users_df.pivot(index='time_hour', columns='bs', values='users')
store = HDFStore('data/users_data.h5')
store.put('users', users_pivot, format='table', data_columns=True)
store.close()
```

```python
packets_df = clustered_traffic_df[['bs', 'time_hour', 'packets']]
packets_pivot = packets_df.pivot(index='time_hour', columns='bs', values='packets')
store = HDFStore('data/packets_data.h5')
store.put('packets', packets_pivot, format='table', data_columns=True)
store.close()
```
