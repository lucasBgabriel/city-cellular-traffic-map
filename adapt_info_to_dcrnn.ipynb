```python
clustered_traffic_df['time_hour'] = clustered_traffic_df['time_hour'].map(lambda x: pd.Timestamp(x, unit='s', tz='Asia/Shanghai'))
clustered_traffic_df['hour'] = clustered_traffic_df['time_hour'].map(lambda x: x.hour)
clustered_traffic_df['day'] = clustered_traffic_df['time_hour'].map(lambda x: x.day)
```

```python
clustered_traffic_df["type"] = clustered_traffic_df["bs"].apply(lambda x: x.split("_")[0])
clustered_traffic_df["fog"] = clustered_traffic_df["bs"].apply(lambda x: int(x.split("_")[1])*10)
clustered_traffic_df["bs"] = clustered_traffic_df["bs"].apply(lambda x: int(x.split("_")[1])*10 + int(x.split("_")[2]) + 1 if x.split("_")[0] != "fog" else int(x.split("_")[1])*10)
```

```python
users_df = clustered_traffic_df[['bs', 'time_hour', 'users']]
users_pivot = users_df.pivot(index='time_hour', columns='bs', values='users')
```

```python
users_pivot
```

```python
store = HDFStore('data/users_data.h5')
```

```python
store.put('users', users_pivot, format='table')
```

```python
store.close()
```

```python
packets_df = clustered_traffic_df[['bs', 'time_hour', 'packets']]
packets_pivot = packets_df.pivot(index='time_hour', columns='bs', values='packets')
store = HDFStore('data/packets_data.h5')
store.put('packets', packets_pivot, format='table')
store.close()
```

```python
bytes_df = clustered_traffic_df[['bs', 'time_hour', 'bytes']]
bytes_pivot = bytes_df.pivot(index='time_hour', columns='bs', values='bytes')
store = HDFStore('data/bytes_data.h5')
store.put('bytes', bytes_pivot, format='table', data_columns=True)
store.close()
```
